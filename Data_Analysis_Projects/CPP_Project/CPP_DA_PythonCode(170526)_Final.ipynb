```python
# Total & Average Spending by 'Age Group'
# Total spending on Previous Purchases by 'Top-tier Customers' by 'Age Group'.

prem111=pd.DataFrame(prem_cust.loc[prem_cust['customer_no'].isin(dfcpp.index)].
                   groupby('age_group',observed=True)['dlr_spent_multi_purch'].sum().sort_values(ascending=True).head(100))
print('Total spending on Previous Purchases by "Top-tier Customers" by "Age Group"')
prem111['Total Amount Spent']=prem111
display(prem111[['Total Amount Spent']])
print()

# Plotting the dataset
prem111['Total Amount Spent'].plot.barh()
plt.xlabel('Total Amount Spent')
plt.ylabel('Age Group')
display('Top-tier Customers Previous Purchases')
plt.title('H-Bar Chart - Total Amount Spent by Age Group')
plt.show(block=False)

print("\033[4mFinding\033[0m:"),
print('''- By Age-group and total amount spent overtime, in the "Top-tier Customers" category, Young Adults (25-35) 
  spent significantly most (more than others combined), followed by Adults (36-65), while Adolescents (10-15) 
  spent the least.''')
print("\033[4mInsight\033[0m:"),    
print('''- Reward and customer loyalty processes should be implemented, high-valued and age-range items
  should always be stocked, to target and further encourage Young Adult customers to maintain their spending 
  patterns, while Adults, Youths (16-24) and Adolecents should be targeted with sales promotion strategies 
  to improve their spending pattern on the Company's items.''')
print()


# Average spending on Previous Purchases by 'Top-tier Customers' by 'Age Group'.

prem112=pd.DataFrame(prem_cust.loc[prem_cust['customer_no'].isin(dfcpp.index)].
                   groupby('age_group',observed=True)['dlr_spent_multi_purch'].mean().sort_values(ascending=True).head(100))
print('Average spending on Previous Purchases by "Top-tier Customers" by "Age Group"')
prem112['Average Amount Spent']=prem112
display(prem112[['Average Amount Spent']])
print()

# Plotting the dataset
prem112['Average Amount Spent'].plot.bar()
plt.xlabel('Age Group')
plt.ylabel('Average Amount Spent')
display('Top-tier Customers Previous Purchases')
plt.title('H-Bar Chart - Average Amount Spent by Age Group')
plt.show(block=False)

print("\033[4mFinding\033[0m:"),
print('''- By Age-group and on average amount spent to previously purchase items, in the "Top-tier Customers" 
  category, Young Adults and Adolescents spent more with an almost equal average amount spent in both age 
  categories. Adults and Youths spent less with also almost equal average amount spent in both age categories.''')
print("\033[4mInsight\033[0m:"),
print('''- All customers (especially Adults and Youths) must be encouraged with top-notch customer relation 
  and sale improvement strategies, to maintain continuously and even increase their spending levels on the 
  Company's items''')
print()
```
